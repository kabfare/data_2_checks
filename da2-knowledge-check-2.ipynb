```python
import pandas as pd
```

### Cleaning Column Names

```python
df = pd.read_csv('happiness_dataset.csv')

print(df.shape)
df
```

```python
# Checking column names

df.columns
```

```python
# Replacing column names with snake case

columns_re = {
    'Overall rank_2019' : 'rank_2019',
    'Country or region' : 'country_region',
    'Score_2019' : 'score_2019',
    'Rank_2018' : 'rank_2018',
    'Score_2018' : 'score_2018',
    'Rank_2017' : 'rank_2017',
    'Score_2017' : 'score_2017',
    ' Rank_2016' : 'rank_2016',
    'Score_2016' : 'score_2016',
    '2020.Rank' : 'rank_2020',
    '2020_Score' : 'score_2020',
    'Rank_2022' : 'rank_2022',
    'Score_2022' : 'score_2022',
    '2021. score' : 'score_2021',
    '2021.Rank' : 'rank_2021'
}

df.rename(columns = columns_re, inplace = True)

df.columns
```

```python
# Changing order of column names to year order so that they make actual sense

df = df[['country_region', 'rank_2016', 'score_2016', 'rank_2017', 
         'score_2017', 'rank_2018', 'score_2018', 'rank_2019', 
         'score_2019', 'rank_2020', 'score_2020', 'rank_2021', 
         'score_2021', 'rank_2022', 'score_2022']]

df.columns
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Changing rank column types to ints for aesthetic purposes

rank_type=[col for col in df.columns if 'rank' in col]

df[rank_type] = df[rank_type].astype('Int64')

df.dtypes
```

```python
df.head()
```

```python
# Replacing NaNs with 0

df.fillna(value=0, inplace=True)
df.tail()
```
