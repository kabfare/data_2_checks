```python
import pandas as pd
import requests as re
import json
```

```python
#pull data from an API

api_url = "https://api.jikan.moe/v4/anime?q=Naruto Shippuden&sfw"
response = re.get(api_url)
response_json = response.json()

#reading the json into a table and filtering by 'data'
df = pd.json_normalize(response_json['data'])
df
```

```python
#checking columns for fun
df.info()
```
